### **1.Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url='https://raw.githubusercontent.com/prasertcbs/basic-dataset/refs/heads/master/store_data.csv'
df = pd.read_csv(url)
df.columns = [f'col{i}' for i in range(df.shape[1])]
df
```

```python
df.info()
```

### **2.แสดงค่าที่ไม่ซ้ำกันใน columns**

```python
for c in df.columns:
  print(f'{c} : {df[c].unique}')
```

### **3.แปลงข้อมูลให้อยู่ในรูปแบบ one-hot Encoder (True/False หรือ 1/0)**

```python
from mlxtend.preprocessing import TransactionEncoder

dataset = df.values.tolist()
dataset[:5]
```

```python
te = TransactionEncoder()
dataset = df.fillna('nan').values.tolist()
te_ary = te.fit(dataset).transform(dataset) #เรียนรู้สินค้าที่ไม่ซ้ำกันทั้งหมด จาก dataset เพื่อสร้าง columns
#ถ้ามีสินค้าใน Transaction เป็น True ไม่มีเป็น False
df_ohe = pd.DataFrame(te_ary, columns=te.columns_)
df_ohe
```

### **4.นำเข้า Apriori Algorithm**

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df_ohe, min_support=0.01, use_colnames=True)
#support (ความถี่ในการปรากฏ)
frequent_itemsets.sort_values('support', ascending=False)
```

### **5.นำเข้า Association Rule**

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
#metric = 'lift' การกรองกฏโดยใช้ lift เป็นตัวชี้วัดบอกความน่าจะเป็นในการซื้อสินค้า
#เพิ่มขึ้นเท่าใดเมื่อ lift > 1 แสดงว่ามีการซื้อสินค้าร่วมกันบ่อย
#lift = 1 แสดงว่าไม่มีความสัมพันธ์
#lift < 1 แสดงว่ามีการซื้อสินค้าร่วมกันน้อย
rules.sort_values('lift', ascending=False)
rules
```

- antecedents (ส่วนที่นำไปสู่)
- consequents (ส่วนที่เป็นผล)
- Spport : ความถี่ของการปรากฏของทั้ง antecedents และ consequents ร่วมกัน
- Confidence : ความน่าจะเป็นที่จะซื้อconsequentsเมื่อซื้อ antecedents
