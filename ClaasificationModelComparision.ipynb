### **1.Import Libary**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print(f'pandas version: {pd.__version__}')
print(f'numpy version: {np.__version__}')
print(f'seaborn version: {sns.__version__}')
print(f'plt version: {plt.rcParams["figure.figsize"]}')
```

### **2.โหลดข้อมูล Iris Database**

```python
from sklearn.datasets import load_iris
iris = load_iris()
print(iris)
```

### **3.กำหนดตัวแปรอิสระ(X) และตัวแปรตาม(y)**

```python
x = pd.DataFrame(iris['data'],columns=iris['feature_names'])
y = pd.Series(data=iris['target'],name='target')
```

### **4.แบ่งข้อมูล Train และ Test**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=7,stratify=y)
#stratify = y สำคัญสำหรับ Classification พารามิเตอร์จะทพให้แบ่งข้อมูลแบบสุ่มข้อมูลยังคงรักษาสัดส่วนของจำนวนของแต่ละคลาส
#ช่วยเมื่อข้อมูลมีคลาวที่ไม่สมดุล (Imbalance classes)
```

### **5.สร้าง list ของ Classifier models**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

model=[]

model.append(('LogisticRegression',LogisticRegression(random_state=7,solver='liblinear')))
model.append(('KNeighborsClassifier',KNeighborsClassifier()))
model.append(('GaussianNB',GaussianNB()))
model.append(('DecisionTreeClassifier',DecisionTreeClassifier(random_state=7)))
model.append(('SVC',SVC(random_state=7)))
model.append(('RandomForestClassifier',RandomForestClassifier(random_state=7)))
model.append(('AdaBoostClassifier',AdaBoostClassifier(random_state=7)))
model.append(('GradientBoostingClassifier',GradientBoostingClassifier(random_state=7)))
```

### **6.ฝึกและประเมินโมเดล**

```python
from sklearn.metrics import accuracy_score
names=[]
results=[]
for name, model_instance in model:
  model_instance.fit(x_train,y_train)  #ฝึกโมเดลด้วยชุดฝึก(x_train, y_train)
  y_pred=model_instance.predict(x_test)  #ใช้โมเดลที่ฝึกทำนายข้อมูลชุดทดสอบ เก็บผลลัพธ์ที่ y_pred
  accuracy=accuracy_score(y_test,y_pred)  #คำนวณค่าความถูกต้อง โดยเปรียบเทียบค่าจริง (y_test) กับค่าที่ทำนายได้ (y_pred)
  names.append(name)
  results.append(accuracy)

  print(f'{name:30} Accuracy: {accuracy:.4f}') #ชิดซ้าย 30 ตัวอักษร ความกว้าง 4 ช่องทศนิยม
```

### **7.สร้าง DataFrame แสดงผลลัพธ์**

```python
dacc = pd.DataFrame(data={'Classifier':names,'Accuracy':results})
dacc.sort_values(by='Accuracy',ascending=False,inplace=True)
dacc.set_index('Classifier',inplace=True)
dacc
```
