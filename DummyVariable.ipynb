DummyVariable คือตัวแปรหุ่น หรือ One-hot Encoding แปลงข้อความเป็นตัวเลข

1.import libary and  make data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#sample data
data = {'age':[25,30,35,40,28],'gender':['M','F','M','F','M'],'country':['THA','CAM','THA','JPN','CAM']}
df = pd.DataFrame(data)
df.head()
df.info()
```

2. สร้าง Dummy variable โดยไม่ทิ้งคอลัมภ์แรก

```python
pd.get_dummies(df['gender'])
```

```python
pd.get_dummies(df['gender'],drop_first=True)
#ลบคลอลัม DummyVariable อันแรกทิ้ง เหลือแค่ M
#drop_first=True ตีความง่ายขึ้น
```

3.สร้าง Dummy Variable ของ County

```python
pd.get_dummies(df['country'])
```

```python
pd.get_dummies(df['country'],drop_first=True)
```

4.รวม Dummy variable เข้ากับ dataFrameหลัก

```python
pd.get_dummies(df,columns=['gender','country'],drop_first=True)
```
