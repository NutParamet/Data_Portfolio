ข้อมูลสูญหาย(Missing Values)
- การเติมข้อมูล(Imputation)
- การแทนที่(Replace)

1.Import libary and load data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer

#load data
df = pd.read_csv('https://raw.githubusercontent.com/zief0002/bespectacled-antelope/main/data/mammals.csv')
df.head()
```

```python
df.info()
```

2.ตรวจสอบจำนวน Missing Values ในแต่ละคอลัม

```python
df.isnull().sum()
```

3.เติม missing values ค่าเฉลี่ย(mean)

```python
df['slow_wave'].fillna(df['slow_wave'].mean(), inplace=True)
#df[slow_wave]=df['slow_wave'].fillna(df['slow_wave'].mean())
df.isnull().sum()
```

4.เติม missing values ด้วยค่ามัธยฐาน(median)

```python
df['paradox'].fillna(df['paradox'].median(), inplace=True)
df.isnull().sum()
```

5.เติม missing values ด้วยค่าฐานนิยม(mode)

```python
df['total_sleep'].fillna(df['total_sleep'].mode()[0], inplace=True) #เลือกค่าที่พบบ่อย คือ index[0]
df.isnull().sum()
```

6.เติม Missing Values ด้วย SimpleImputer

```python
imp_mean = SimpleImputer(strategy='mean') #กำหนดกลยุทธ์
imp_mean.fit_transform(df[['lifespan']]) #2D array
```

```python
imp_median = SimpleImputer(strategy='median') #กำหนดกลยุทธ์
imp_median.fit_transform(df[['lifespan']]) #2D array
```

```python
imp_mode = SimpleImputer(strategy='most_frequent') #กำหนดกลยุทธ์
imp_mode.fit_transform(df[['lifespan']]) #2D array
```

```python
df.isnull().sum()
```

7.เติมค่าSimpleimputerในdataframe

```python
df['lifespan'] = imp_mean
df.isnull().sum()
```

```python
df['gestation'] = imp_median
df.isnull().sum()
```
